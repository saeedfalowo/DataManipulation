# Extracting Car Data from a .csv file

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
csvfilepath = 'CarData/Car_Pzz.csv'
content = []
content_tuple = ()

with open(csvfilepath, newline='') as csvfile:
    csv_read = csv.reader(csvfile,delimiter=',',quoting=csv.QUOTE_NONNUMERIC)
    #SigNames = next(csv_read)
    #data = np.array(list(csv_read))#.dtype(float)
    for row in csv_read:
        #print(', '.join(row))
        content.append(row)
        #print(row)

#content_tuple = content
#print(content)
```

```python
content
```

```python
numrows = len(content)
numcols  = len(content[0])

print('content consists of %d rows and %d columns' %(numrows,numcols))
```

```python
signames = content[0]
dataset = content[1:numrows]
#dataset_f = map(float, dataset)
#dataset_f
type(dataset)
```

```python
df = pd.DataFrame(dataset)
#df = df.astype(float)
# find empty and nan cells
print('These are the row and column indices that has with empty cells:')
print(np.where(pd.isnull(df)))
#for i in range(numcols):
    #df[i] = df[i].astype(float)
```

```python
# It seems like row 46646, which is the last row contain empty cells for all the columns
# remove the last row from the dataFrame
dflastrow = len(df)-1
df = df.drop([dflastrow], axis=0)
```
